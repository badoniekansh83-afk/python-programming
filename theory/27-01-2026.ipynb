## Wap to find a number whether it is palindrome or not eg 121####

```python
a=121
temp=a
rev=0
while a>0:
    b=a%10
    rev=rev*10 +b
    a=a//10
if rev==temp:
    print("It is palindrome")
else:
    print("It is not palindrome")
```

1. firstly we input a value of a=121 and create a temporary variable as temp and store it as a
2. we also take a variable as rev to reverse that number
3. now we use while loop in a>0 and we take a remainder by using a variable as b
4. now we multiplle rev by 10 and add the remainder b after removing by 10 until the condition turn by false
5. now we check if rev is equal to temp then print it is a palindrome or not then print not a palindrome

## Wap to find whether it a armstrong or not 

```python
a=153
temp=a
n=len(str(a))
b=0
while int(a)>0:
    r=a%10
    b=b+(r**n)
    a=a//10
if temp==b:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

1. a = 153 → we have to check
2. n = len(str(a)) → counts how many digits are in the number
3. The while loop extracts each digit (r) one by one
4. Each digit is raised to the power n and added to b
5. Finally, it compares the original number (temp) with the calculated sum (b)
6. If both are equal, it prints “Armstrong Number”, otherwise “Not an Armstrong Number”
